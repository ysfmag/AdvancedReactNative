```
import os

CONST_FIRST = 'Rapports_verification_ACITOP_LOT2(234)_'


def rename_rapport_file():
    for count, filename in enumerate(os.listdir("./rapports")):
        dst = filename.replace(CONST_FIRST, '')
        src = './rapports/' + filename
        dst = './rapports/' + dst
        os.rename(src, dst)


def transfer_xlsx_file_into_directory():
    for count, filename in enumerate(os.listdir("./rapports")):
        if ".xlsx" in filename:
            dst = filename.replace('.xlsx', '')
            src = './rapports/' + filename
            dst = './rapports/' + dst + "/" + filename
            os.rename(src, dst)

```

```
print('start rename all file')
rename_rapport_file()
print('move .xlsx file into directory')
transfer_xlsx_file_into_directory()
```
